# KMeans

Random đầu vào

```python
def random_init_centroids(X, K, random_state = None):
    np.random.seed(random_state) 
    centroids = np.zeros((K, X.shape[1]))
    
    idx = np.random.permutation(X.shape[0]) # random ngẫu nhiên các index
    centroids = X[idx[:K], :] # lấy ra các điểm trong tập X
    
    return centroids 
```

Gán các điểm cho centroids gần nó nhất
* Khoảng các sẽ được tính theo công thức Euclid

```python
def find_closest_centroids(X, centroids):   
    K = centroids.shape[0]
    idx = np.zeros((X.shape[0], 1), dtype = np.int8)
    
    for i in range(X.shape[0]): 
        min_distance = 0 # lưu khoảng cách bé nhất của một điểm đang xét đến centroids
        
        for j in range(K): 
            diff = X[i, :] - centroids[j, :] 
            distance = np.sum(np.power(diff, 2)) 
            
            if(j == 0 or distance < min_distance):
                min_distance = distance
                idx[i] = j # gán nhãn cho điểm i
                
    return idx      
```

Tính toán tọa độ mới của centroids
* Tọa độ của một chiều sẽ bằng trung bình cộng của tất các điểm thuộc centroid đó ứng với chiều tương ứng

```python
def compute_centroids(X, idx, K):
    centroids = np.zeros((K, X.shape[1]))
    
    for i in range(K):
        idx_cluster= np.where(idx == i)[0] # lấy ra index các điểm thuộc cluster i
        centroids[i, :] = np.mean(X[idx_cluster, :], axis = 0) # tính trung bình theo từng trục để tìm tọa độ mới cho centroids
        
    return centroids
```

Một số hàm để plot kết quả

```python
def draw_line(previous_point, current_point, color = None):
    x = [previous_point[0], current_point[0]]
    y = [previous_point[1], current_point[1]]
    
    plt.plot(x, y, c = color, zorder = 2)

def hsv_to_rgb(n = 63): 
    return colors.hsv_to_rgb(np.column_stack([np.linspace(0, 1, n + 1), np.ones((n + 1, 2))]))

def plot_assigned_points(X, idx, K):
    rgb = hsv_to_rgb(n = K)
    
    for i in range(K):
        idx_cluster = np.where(idx == i)[0]
        plt.scatter(X[idx_cluster, 0], X[idx_cluster, 1], s = 50, facecolors = 'none', edgecolors = rgb[i], zorder = 1, label = "cluster " + str(i))
```

KMeans

```python
def run_KMeans(X, K, max_iter, init_centroids = None, draw_progress = True, random_state = None):    
    centroids = init_centroids 
    idx = np.zeros((X.shape[0], 1))
    previous_centroids = np.zeros((K, X.shape[1])) # dùng để vẽ progress
    
    if(centroids is None):
        centroids = random_init_centroids(X = X, K = K, random_state = random_state)
    
    if(draw_progress):
        plt.close()
        plt.figure(figsize = (10, 5))
        plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'o', c = 'k', s = 100, zorder = 2, label = "start")
        previous_centroids = centroids
    
    for i in range(max_iter): 
        idx = find_closest_centroids(X = X, centroids = centroids) # tìm centroid gần nhất tới từng điểm
        centroids = compute_centroids(X = X, idx = idx, K = K) # cập nhập centroids
        
        if(draw_progress):
            plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', c = 'k', s = 50, zorder = 2)
            
            for j in range(K):
                draw_line(previous_point = previous_centroids[j, :], current_point = centroids[j, :], color = 'k')
            
            if(i == (max_iter - 1)):
                plt.scatter(centroids[:, 0], centroids[:, 1], marker = 'x', c = 'k', s = 50, zorder = 2, label = "update")
                plt.scatter(centroids[:, 0], centroids[:, 1], marker = '>', c = 'k', s = 200, zorder = 2, label = "end")
                plot_assigned_points(X = X, idx = idx, K = K)              
                plt.legend(fontsize = 10, loc = "lower center", ncol = 3)
                plt.axis([-3.5, 1.5, 0.5, 3.5])
                plt.show()
                     
            previous_centroids = centroids
                
    return centroids, np.array(idx.ravel())
```

Chạy thử

```python
centroids, idx = run_KMeans(X = X, K = 5, max_iter = 10, random_state = 42)
print(centroids)
print(idx)
```

Việc gán nhãn cho từng điểm như trên gọi là hard clustering
* Trong một số bài toán ta chỉ gần tính khoảng cách của điểm đó đến từng centroids
* Các giá trị đó có thể là weights đầu vào cho các thuật toán khác
* Việc tính toán như trên gọi là soft clustering

```python
def find_distance_to_centroids(X, centroids):
    distance = []
    
    for i in range(X.shape[0]):
        tmp_distance = []
        
        for j in range(centroids.shape[0]):
            diff = X[i, :] - centroids[j, :]
            dis = np.sum(np.power(diff, 2))
            tmp_distance.append(np.sqrt(dis))
            
        distance.append(tmp_distance)
        
    return np.array(distance)
```

# Xử lý Initilizing centroids

Chọn các điểm centroids làm đầu vào cho thuật toán cũng ảnh hưởng rất nhiều đến thuật toán

Nếu như đầu vào tốt thì thuật toán cho ra kết quả tốt, và thậm chí việc converge cũng nhất nhanh

```python
_, _ = run_KMeans(X = X, K = 5, max_iter = 10, random_state = 42)
```

Nếu như đầu vào xấu thì kết quả đầu ra sai và khó có thể phân loại được chính xác các cụm

```python
_, _ = run_KMeans(X = X, K = 5, max_iter = 10, random_state = 8)
```

Để tìm được đầu vào tốt thì có thể dùng đến inertia (tổng khoảng của các đểm đến centroid của nó là bé nhất)

```python
def find_inertia(X, centroids, idx):
    distance = find_distance_to_centroids(X = X, centroids = centroids) # khoảng cách từ từng điểm đến từng centroid
    
    return np.sum(distance[np.arange(len(X)), idx] ** 2)
```

```python
def find_best_init_centroids(X, K, n, print_process = None):
    inertia = -1
    best_init_centroids = np.zeros((K, X.shape[1]))
    
    for i in range(n): 
        init_centroids = random_init_centroids(X = X, K = K)
        centroids, idx = run_KMeans(X = X, K = K, max_iter = 10, init_centroids = init_centroids, draw_progress = False)
        tmp_inertia = find_inertia(X = X, centroids = centroids, idx = idx) 
        
        if(print_process):
            print("i = {}:\ninit_centroids:\n{}".format(i, init_centroids))
            print("inertia: {}\n".format(tmp_inertia))
        
        if(inertia == -1 or tmp_inertia < inertia): # tìm inertia bé nhất
            inertia = tmp_inertia
            best_init_centroids = init_centroids            
        
    return best_init_centroids, inertia
```

```python
best_init_centroids, inertia = find_best_init_centroids(X = X, K = 5, n = 8, print_process = True)
print("best choice:\ninit_centroids:\n{}".format(best_init_centroids))
print("inertia: {}\n".format(inertia))
```

```python
centroids, idx = run_KMeans(X, K = 5, max_iter = 10, init_centroids = best_init_centroids)
```

# Xử lý Initilizing K

Việc chọn số cụm cũng ảnh hưởng rất nhiều đến thuật toán, việc chọn quá nhiều hay quá ít đều mang đến kết quả xấu

```python
c, i = run_KMeans(X, K = 15, max_iter = 10)
```

Với việc chọn k thì không thể dùng thuần inertia được thì về bản chất càng nhiều centroids thì inertia càng bé

```python
find_inertia(X = X, centroids = c, idx = i)
```

Có một phương pháp được dùng kết hợp với inertia là Elbow
* Chạy thuật toán trên một tập giá trị k, sau đó biểu diễn và chọn tại vị trí "khủy tay"
* Chọn ở vị trí này vì từ vị trí "vai" đến "khủy tay" thì inertia vẫn còn có thể giảm nhiều, còn từ "khủy tay" đến cổ tay thì interia giảm rất ít

```python
def find_init_k_using_inertia(X, n):
    inertias = []
    
    for k in range(2, n + 1):
        best_init_centroids, _ = find_best_init_centroids(X = X, K = k, n = 10)
        centroids, idx = run_KMeans(X = X, K = k, max_iter = 10, init_centroids = best_init_centroids, draw_progress = False)
        inertia = find_inertia(X = X, centroids = centroids, idx = idx.ravel())
        inertias.append(inertia)
        
        print("K = {}\nInertia = {}".format(k,  inertia))
        
    return np.array(inertias)
```

```python
inertias = find_init_k_using_inertia(X = X, n = 8)
```

```python
plt.close()

plt.figure(figsize = (10, 5))
plt.plot(range(2, 8 + 1), inertias, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)

plt.show()
```

Nhưng kết quả ở trên lại cho kết quả xấu hơn cả k = 5 - phương pháp tiếp theo là dùng Silhouette score

```python
def find_distance_matrix(X): 
    distance_matrix = []
    
    for i in range(X.shape[0]):
        diff = X[i] - X
        dis = np.sum(np.power(diff, 2), axis = 1)
        distance_matrix.append(np.sqrt(dis))
    
    return np.array(distance_matrix) # trả về ma trận m * m là khoảng cách từ từng điểm đến từng điểm còn lại, với m là số lượng hàng trong X
```

Slihoutte score được tính như sau:  
$\frac{b - a}{max(a, b)}$
* a là khoảng khách trung bình của một điểm đến tất cả các điểm còn lại trong một cụm
* b là khoảng các trung bình của một điểm đến các điểm của cụm gần nó nhất

```python
def find_a(distance_matrix, idx):
    a = []
    
    for i in range(distance_matrix.shape[0]):
        idx_cluster_i = np.where(idx == idx[i])[0]
        average = np.sum(distance_matrix[i, idx_cluster_i]) / (idx_cluster_i.shape[0] - 1)
        a.append(average)

    return np.array(a) # trả về trung bình khoảng cách từ một điểm đến các điểm còn lại thuộc cùng một cluster
```

```python
def find_b(distance_matrix, idx, K):
    b = []
    
    for i in range(distance_matrix.shape[0]):
        average_to_other_cluster = []
        
        for j in range(K):
            if(idx[i] != j):
                idx_cluster_j = np.where(idx == j)[0]
                average = np.sum(distance_matrix[i, idx_cluster_j]) / (idx_cluster_j.shape[0])
                average_to_other_cluster.append(average)
        
        b.append(np.min(average_to_other_cluster)) # lấy min là trung bình khoảng cách bé nhất
        
    return np.array(b) # trả về trung bình khoảng các từ một điểm đến các điểm thuộc cluster khác và gần nhất với điểm đó
```

```python
def find_silhoutte_for_points(a, b): # tính silhoutte cho từng điếm (b - a) / max(a, b)
    return np.array([(b[i] - a[i]) / np.max([a[i], b[i]]) for i in range(a.shape[0])])
```

```python
def find_silhoutte_for_clusters(silhoutte_for_points, idx, K):
    
    silhoutte_for_clusters = []
    for i in range(K):
        idx_cluster_i = np.where(idx == i)[0]
        sil = np.sum(silhoutte_for_points[idx_cluster_i]) / idx_cluster_i.shape[0]
        silhoutte_for_clusters.append(sil)
    
    return np.array(silhoutte_for_clusters) # trả về slihoutte trung bình cho từng cluster
```

```python
def silhoutte_score(X, idx, K, draw_diagram = False):
    distance_matrix = find_distance_matrix(X)
    a = find_a(distance_matrix, idx)
    b = find_b(distance_matrix, idx, K)
    silhoutte_for_points = find_silhoutte_for_points(a, b)
    silhoutte_for_points_clusters = find_silhoutte_for_clusters(silhoutte_for_points, idx, K)
    score =  np.sum(silhoutte_for_points_clusters) / K
    
    print("K = {}\nSilhoutte_score = {}".format(K,  score))
    if(draw_diagram == True):
        padding = len(X) // 30
        pos = padding
        colors = hsv_to_rgb(n = K)
        plt.subplot(420 + K - 1) # k chạy ít nhất từ 2 nên phải trừ đi 1, tính từ 421 trở đi
        plt.yticks([])
        
        for i in range(K):
            coeffs = silhoutte_for_points[np.where(idx == i)[0]]
            coeffs.sort()
            
            plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs, facecolor = colors[i])
            
            pos += len(coeffs) + padding
            
        plt.axvline(score, color = "black", linestyle = "--")
        plt.title("K = {}".format(K))
        
    return score # trả về slihoutte trung bình cho toàn dữ liệu
```

```python
def find_init_k_using_silhoutte(X, n, draw_diagram = False):
    silhoutte_scores = []
    
    plt.close()
    plt.figure(figsize = (10, 10))
    plt.subplots_adjust(wspace = 0.5, hspace = 0.5)

    
    for k in range(2, n + 1):
        best_init_centroids, _ = find_best_init_centroids(X = X, K = k, n = 10)
        centroids, idx = run_KMeans(X = X, K = k, max_iter = 10, init_centroids = best_init_centroids, draw_progress = False)
        score = silhoutte_score(X = X, idx = idx.ravel(), K = k, draw_diagram = draw_diagram)
        silhoutte_scores.append(score)
        
    return np.array(silhoutte_scores)
```

```python
silhoutte_scores = find_init_k_using_silhoutte(X = X, n = 8, draw_diagram = True)
plt.show()
```

Trong hình trên trục y là chiều chỉ các cụm, trục x là silhouetee score, ta thấy với k = 5 thì các cụm có silhouette vượt qua trung bình và các cluster có kích thước khá đều
